```python
charts_df = pd.read_csv('data/charts.csv', sep=';')
charts_df.head()
```

```python
#To analyze, Let's merge first using outer join
df_outermerged = pd.merge(charts_df, df_songs, on='song_id', how='outer')

# Rows containing only information from the chatts_df data set
only_charts = df_outermerged[df_outermerged['CHART_ID'].notna() & df_outermerged['artist_name'].isna()].shape[0]

# # Rows containing only information from the df_songs data set
only_songs = df_outermerged[df_outermerged['artist_name'].notna() & df_outermerged['CHART_ID'].isna()].shape[0]


# Print results

print("Number of rows containing only information from the CHARTS data set with NO Corrosponding song metadata:", only_charts)
print("Number of rows containing only information from the SONGS data set with No Corrosponding Charts metadata:", only_songs)
```

Based on the results, 53 rows in the charts.csv dataset have no corresponding metadata in the songs.csv dataset, while 376 rows in songs.csv do not appear in charts.csv. This indicates potential discrepancies between the two datasets, such as missing or unmatched song_id values. To ensure no information is lost, an outer merge is the most suitable approach, as it retains all rows from both datasets, filling missing values with NaN where necessary. This approach allows for comprehensive analysis while identifying and handling unmatched data effectively.

```python
df_outermerged = pd.merge(charts_df, df_songs, on='song_id', how='outer')
df_outermerged.shape
```

## Exercise 7 (15 points)

Consider the charts of the week prior to christmas (date: 2021-12-23). How many songs that appeared in the German (de) charts during this week also appear in the charts of the Unites States (us)?

```python
df_outermerged.head()
```

```python
german = df_outermerged[(df_outermerged['DATE'] <= '2021-12-23') & (df_outermerged['COUNTRY'] == 'de')]
usa = df_outermerged[(df_outermerged['DATE'] <= '2021-12-23') & (df_outermerged['COUNTRY'] == 'us')]

# Find the intersection of song IDs between the German and US charts
common_songs = set(german['song_id']).intersection(set(usa['song_id']))

# Count the number of common songs
num_common_songs = len(common_songs)

# Print the result
print(f"Number of songs in the German charts that also appear in the US charts on 2021-12-23: {num_common_songs}")
```

```python
# Filter the DataFrame for the week prior to Christmas (2021-12-23) and German charts
german_charts = df_outermerged[(df_outermerged['DATE'] == '2021-12-23') & (df_outermerged['COUNTRY'] == 'de')]

# Filter the DataFrame for the week prior to Christmas (2021-12-23) and US charts
us_charts = df_outermerged[(df_outermerged['DATE'] == '2021-12-23') & (df_outermerged['COUNTRY'] == 'us')]

# Merge the two DataFrames on 'song_id' to find common songs
common_songs = pd.merge(german_charts, us_charts, on='song_id', how='inner')

# Count the number of common songs
num_common_songs = common_songs['song_id'].nunique()

# Print the result
print(f"Number of songs in the German charts that also appear in the US charts on 2021-12-23: {num_common_songs}")
```

```python
df_outermerged = df_outermerged.drop(columns= ['TYPE', 'COUNTRY', 'INTERVAL', 'DATE'], axis=1)
```
