````python
import requests
import pandas as pd
import json


def fetch(url, params={}):
    response = requests.get(url,params)
    data = response.json()
    return data

def fetchAndDf(url, params={}):
    _data = fetch(url, params)
    _df = pd.DataFrame(_data['data'])
    _df.columns = _data['fields']
    return _df
````
