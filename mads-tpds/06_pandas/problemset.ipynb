```python
import pandas as pd
```

```python
def format_excel(df):
    df_ref = df.transpose()
    df_ref.columns = df_ref.iloc[0]
    df_ref = df_ref.drop(df_ref.index[0])

    df_ref.set_index('track_id', inplace=True)
    # print(df_ref.index)
    # df_ref.drop(columns='variable', inplace=True)
    # df_ref.set_index('')
    return df_ref


def format_excels(paths):
    list = []
    for path in paths:
        df = pd.read_excel(path)
        country = path.split('/')[-1].split('.')[0].split('_')[1].upper()
        fdf = format_excel(df)
        fdf['country'] = country
        list.append(fdf)
    return list
```

```python
dfs_charts = format_excels(['data/charts_es.xlsx', 'data/charts_de.xlsx', 'data/charts_us.xlsx', 'data/charts_gb.xlsx'])
df_charts = pd.concat(objs=dfs_charts)
df_charts
```

```python
df_lyrics = pd.read_csv('data/lyrics.csv')
df_lyrics.set_index('id', inplace=True)
df_lyrics
```

```python
df_tracks = pd.read_csv('data/tracks.csv')
df_tracks.set_index('id', inplace=True)
df_tracks
```

```python
df_tracks_lyrics = pd.merge(left=df_tracks, right=df_lyrics, how='left', left_index=True, right_index=True)
df_tracks_lyrics
```

```python
df_tracks_lyrics_charts = pd.merge(left=df_tracks_lyrics, right=df_charts, how='outer', left_index=True, right_index=True)
df_tracks_lyrics_charts
```

```python
df_artists = pd.read_json('data/artists.json')
df_artists.set_index('id', inplace=True)
df_artists
```

```python
df = pd.merge(left=df_tracks_lyrics_charts, right=df_artists, how='left', left_on='artist_ids', right_index=True)
df
```
