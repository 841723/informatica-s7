```python
def search4patterns(message,size):
    l = {}
    for i in range(0,len(message)-size+1):
        pattern2look = message[i:i+size]
        for newi in range(i+size,len(message)-size+1):
            p = message[newi:newi+size]
            if (p==pattern2look):
                if p in l:
                    l[p].occurence += 1
                    l[p].positions.append([i,newi])
                else:
                    l[p] = {'ocurence': 1, 'positions': [[i,newi]]}

    return l            
            

l = search4patterns(ciphertext,5)
l
```
